### MAking changing to the data 

```python
df.head(3)
```

```python
df['Total']  = df['HP'] + df['Attack']  + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed'] 
#df['Total']
df.head(5)
```

```python
45	+ 49	+49	+65	+65+	45
```

```python


df = df.drop(columns=['Total'])
df.head(3)
```

```python
df['Total'] = df.loc[:, 'HP':'Speed'].sum(axis=1)
```

```python
df
```

```python
# Custom column order
cols = list(df.columns)

# Remove 'Total' and insert after 'Name'
cols.remove('Total')
cols.insert(cols.index('Name') + 1, 'Total')

# Reorder the DataFrame
df = df[cols]
```

```python
df
```

### SAving our data into Excel , CSV

### Aggregate statistics (Group by)

```python
df.groupby('Type 1')[['HP', 'Attack', 'Defense', 'Total']].mean()
```

```python
df.groupby('Type 1').size()
# OR
df['Type 1'].value_counts()
```

```python
df.groupby('Type 1')['Total'].mean().sort_values(ascending=False)
```

```python
grouped_df = df.groupby('Type 1')[['Total']].mean().reset_index()
```
